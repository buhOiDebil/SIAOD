```python
n, S = map(int, input().split())
a = list(map(int, input().split()))


def Prices(a, n, k):
    b = []
    result = 0
    for i in range(n):
        temp = a[i]+(i+1)*k
        b.append(temp)
    for j in range(k):
        result += b[j]
    return result

l = 0
r = n+1
while (l < r-1):
    m = (l+r)//2
    if Prices(a,n,m) <= S:
        l = m
    else:
        r = m


print(l, Prices(a,n,l))
```

–í –ø–µ—Ä–≤–æ–º –ø—Ä–∏–º–µ—Ä–µ –±—ã–ª–æ –∑–∞–≥–∞–¥–∞–Ω–æ —á–∏—Å–ª–æ ùëõ=3, —Ç–æ–≥–¥–∞ ùëõ‚ãÖ4+6‚àí5=13=ùëö. –¢–∞–∫ –∂–µ –≤–æ–∑–º–æ–∂–Ω—ã —Ç–∞–∫–∏–µ –æ—Ç–≤–µ—Ç—ã: ùëé=4, ùëè=5, ùëê=4 (–ø—Ä–∏ —ç—Ç–æ–º ùëõ=3); ùëé=5, ùëè=4, ùëê=6 (–ø—Ä–∏ —ç—Ç–æ–º ùëõ=3); ùëé=6, ùëè=6, ùëê=5 (–ø—Ä–∏ —ç—Ç–æ–º ùëõ=2); ùëé=6, ùëè=5, ùëê=4 (–ø—Ä–∏ —ç—Ç–æ–º ùëõ=2).

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –±—ã–ª–æ –∑–∞–≥–∞–¥–∞–Ω–æ —á–∏—Å–ª–æ ùëõ=1, —Ç–æ–≥–¥–∞ ùëõ‚ãÖ2+2‚àí3=1=ùëö. –ß–∏—Å–ª–æ ùëõ=0 –Ω–µ –º–æ–≥–ª–æ –±—ã—Ç—å –∑–∞–≥–∞–¥–∞–Ω–æ, —Ç–∞–∫ –∫–∞–∫ —á–∏—Å–ª–æ ùëõ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–º.

```python
def check(a, b, c, m):
    if m - (b - c) != 0:
        if (m - (b - c)) % a == 0:
            print(a, b, c)
            return True
    b, c = c, b
    if m - (b - c) != 0:
        if m - (b - c) % a == 0:
            print(a, b, c)
            return True
    return False


def solve():
    l, r, m = map(int, input().split())
    idfk = r - l
    for i in range(l, r+1):
        a = i
        md = m % a
        if md <= idfk:
            b = r
            c = r - md
            if check(a, b, c, m):
                return
        if a-md <= idfk:
            b = r
            c = r - (a - md)
            if check(a, b, c, m):
                return


t = int(input())
for i in range(t):
    solve()
```

```python
def isOK(res, d):
    sum = res * (res + 1) / 2
    if sum < d:
        return False
    return sum % 2 == d % 2


t = int(input())

for i in range(t):
    a, b = map(int, input().split())
    d = abs(a - b)
    if d == 0:
        print("0")
        continue
    res = 1
    while not isOK(res, d):
        res += 1
print(res)
```
